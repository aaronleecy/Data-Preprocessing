# Simple Linear Regression

Data with two (2) variables.

1. Years of experience
2. Salary

Problem : What is the correlation between the two (2) varaibles?

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing the dataset
dataset = pd.read_csv('Salary_Data.csv')

dataset

# 30 observations
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

```python
# Fitting the linear regression on the training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predicting the Test set 
y_pred = regressor.predict(X_test)
```

```python
# Lets compare y_pred predicted values for y against the test dataset.
pd.DataFrame(y_pred)
```

```python
pd.DataFrame(y_test)
```

```python
# Visualizing the Training set 
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

# Red points are actual salary
# Blue line is the linear regression line 

# Will use the same linear regression line to predict the test set
```

```python
# Visualizing the Test set 
plt.scatter(X_test, y_test, color = 'red')

# Note: We still use the trained data 
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
